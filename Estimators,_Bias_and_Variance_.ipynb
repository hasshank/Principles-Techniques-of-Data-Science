```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
p = 0.5
coin_df = pd.DataFrame({"x":[1,0],"p(X=x)":[p,1-p]})
coin_df
```

```python
coin_df.sample(10,weights= "p(X=x)", replace=True)["x"]
```

```python
N=10000
np.random.rand(N,2) < p
```

```python
sim_flips = pd.DataFrame({"Coice A": np.sum((np.random.rand(N,2) < p) * 10, axis=1)})
sim_flips
```

```python
sim_flips["Choice B"] = np.sum((np.random.rand(N,20) < p), axis=1)
sim_flips
```

```python
px.histogram(sim_flips.melt(), x="value", facet_row="variable",barmode="overlay",histnorm="probability",title="empirical distributions", width=600, height=600)
```

```python
pd.DataFrame([
  sim_flips.mean().rename("Simulated Mean"),
  sim_flips.var().rename("Simulated Var"),
  np.sqrt(sim_flips.var()).rename("Simulated SD")])
```
